```python
LATENT_DIM = 64
B = 0.1                 # support length (acts like beta)
b_m = 100 ** (-2/9)     # step size ≈ 0.3594
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np

# ===== ADDED =====
from torchvision.models import inception_v3
from scipy import linalg
# =================


def show_reconstructions(model, loader, device, epoch, n=8):
    model.eval()
    with torch.no_grad():
        x, _ = next(iter(loader))
        x = x.to(device)[:n]
        x_hat, _ = model(x)

    x = x.cpu()
    x_hat = x_hat.cpu()

    fig, axes = plt.subplots(2, n, figsize=(2*n, 4))
    for i in range(n):
        axes[0, i].imshow(x[i, 0], cmap="gray")
        axes[0, i].axis("off")
        axes[1, i].imshow(x_hat[i, 0], cmap="gray")
        axes[1, i].axis("off")

    axes[0, 0].set_ylabel("Input")
    axes[1, 0].set_ylabel("Recon")
    plt.suptitle(f"Epoch {epoch}")
    plt.show()


def sample_epanechnikov(shape, device):
    u1 = torch.rand(shape, device=device) * 2 - 1
    u2 = torch.rand(shape, device=device) * 2 - 1
    u3 = torch.rand(shape, device=device) * 2 - 1
    return torch.median(torch.stack([u1, u2, u3]), dim=0).values


class Encoder(nn.Module):
    def __init__(self, latent_dim):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(1, 32, 4, 2, 1),
            nn.BatchNorm2d(32),
            nn.ReLU(),

            nn.Conv2d(32, 64, 4, 2, 1),
            nn.BatchNorm2d(64),
            nn.ReLU(),

            nn.Conv2d(64, 128, 4, 2, 1),
            nn.BatchNorm2d(128),
            nn.ReLU(),

            nn.Conv2d(128, 256, 4, 2, 1),
            nn.BatchNorm2d(256),
            nn.ReLU(),

            nn.AdaptiveAvgPool2d(1)
        )

        self.fc_mu = nn.Linear(256, latent_dim)
        self.fc_r = nn.Linear(256, latent_dim)

    def forward(self, x):
        h = self.conv(x)
        h = h.view(h.size(0), -1)
        mu = self.fc_mu(h)
        r = F.softplus(self.fc_r(h)) + 1e-6
        return mu, r


class Decoder(nn.Module):
    def __init__(self, latent_dim):
        super().__init__()
        self.fc = nn.Linear(latent_dim, 256)

        self.deconv = nn.Sequential(
            nn.ConvTranspose2d(256, 128, 4, 2, 1),
            nn.BatchNorm2d(128),
            nn.ReLU(),

            nn.ConvTranspose2d(128, 64, 4, 2, 1),
            nn.BatchNorm2d(64),
            nn.ReLU(),

            nn.ConvTranspose2d(64, 32, 4, 2, 1),
            nn.BatchNorm2d(32),
            nn.ReLU(),

            nn.ConvTranspose2d(32, 1, 4, 2, 1),
            nn.Sigmoid()
        )

        self.final = nn.ConvTranspose2d(1, 1, 4, 2, 1)

    def forward(self, z):
        h = self.fc(z)
        h = h.view(-1, 256, 1, 1)
        x_hat = self.deconv(h)
        x_hat = self.final(x_hat)
        x_hat = torch.sigmoid(x_hat)
        return x_hat


class EVAE(nn.Module):
    def __init__(self, latent_dim, B=0.1, b_m=0.3594):
        super().__init__()
        self.encoder = Encoder(latent_dim)
        self.decoder = Decoder(latent_dim)
        self.B = B
        self.b_m = b_m

    def reparameterize(self, mu, r):
        device = mu.device
        K = sample_epanechnikov(mu.shape, device)
        Z = mu + r * K
        U = (torch.rand_like(Z) - 0.5) * self.B
        return self.b_m * Z + U

    def forward(self, x):
        mu, r = self.encoder(x)
        z = self.reparameterize(mu, r)
        x_hat = self.decoder(z)
        return x_hat, r


def evae_loss(x, x_hat, r, B):
    M = x.size(0)
    recon = F.binary_cross_entropy(x_hat, x, reduction='sum') / M
    reg = (3 * B / (5 * M)) * torch.sum(1.0 / r)
    return recon + reg, recon, reg


from torchvision import datasets, transforms
from torch.utils.data import DataLoader

device = "cuda" if torch.cuda.is_available() else "cpu"

transform = transforms.Compose([
    transforms.Resize(32),
    transforms.ToTensor()
])

train_ds = datasets.FashionMNIST(
    root="./data", train=True, download=True, transform=transform
)
loader = DataLoader(train_ds, batch_size=128, shuffle=True)

model = EVAE(latent_dim=64, B=0.1).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)


# ===== ADDED: FID + SHARPNESS =====

class InceptionFeatureExtractor(nn.Module):
    def __init__(self, device):
        super().__init__()
        self.net = inception_v3(pretrained=True)  # FIX
        self.net.fc = nn.Identity()
        self.net.eval().to(device)

    @torch.no_grad()
    def forward(self, x):
        x = x.repeat(1, 3, 1, 1)
        x = F.interpolate(x, size=(299, 299), mode='bilinear', align_corners=False)
        return self.net(x)

def compute_fid(real_feats, fake_feats):
    mu1, mu2 = real_feats.mean(0), fake_feats.mean(0)
    sigma1 = np.cov(real_feats, rowvar=False)
    sigma2 = np.cov(fake_feats, rowvar=False)

    diff = mu1 - mu2
    covmean, _ = linalg.sqrtm(sigma1 @ sigma2, disp=False)
    if np.iscomplexobj(covmean):
        covmean = covmean.real

    return diff @ diff + np.trace(sigma1 + sigma2 - 2 * covmean)


def compute_sharpness(x):
    dx = x[:, :, :, 1:] - x[:, :, :, :-1]
    dy = x[:, :, 1:, :] - x[:, :, :-1, :]
    return (dx.abs().mean() + dy.abs().mean()).item()


fid_extractor = InceptionFeatureExtractor(device)
# ==================================


from tqdm import tqdm

for epoch in range(51):
    model.train()
    total = 0

    for x, _ in tqdm(loader, desc=f"Epoch {epoch}"):
        x = x.to(device)
        optimizer.zero_grad()
        x_hat, r = model(x)
        loss, recon, reg = evae_loss(x, x_hat, r, model.B)
        loss.backward()
        optimizer.step()
        total += loss.item()

    # ===== ADDED METRICS =====
    if epoch % 5 == 0:
        model.eval()
        with torch.no_grad():
            x_real, _ = next(iter(loader))
            x_real = x_real.to(device)[:128]
            x_fake, _ = model(x_real)

            real_feats = fid_extractor(x_real).cpu().numpy()
            fake_feats = fid_extractor(x_fake).cpu().numpy()

            fid = compute_fid(real_feats, fake_feats)
            sharp = compute_sharpness(x_fake)

        show_reconstructions(model, loader, device, epoch)
        print(f"FID: {fid:.2f} | Sharpness: {sharp:.5f}")
    # ========================

    print(f"Epoch {epoch:03d} | Avg Loss {total/len(loader):.4f}")
```
